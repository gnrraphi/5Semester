```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
k = 1
N = 50
e = 1

def S_e(E):
    M = int(E/e)
    if M<0 or M>N:
        return 0
    return k *  np.log(np.math.factorial(N)/np.math.factorial(N-M)/np.math.factorial(M))

def S_a(E):
    M = E/e
    if M<=0 or M>=N:
        return 0
    return k *  (N*np.log(N)-(N-E/e)*np.log(N-E/e)-E/e*np.log(E/e))

def T(E):
    M = E/e
    if M<=0 or M>=N:
        return 0
    T_inv=k/e * (np.log(N-M)-np.log(M))
    if T_inv==0:
        return np.NAN
    return 1/T_inv

def M(T):
    if T==0:
        return np.NAN
    return N/(np.exp((e/(k*T)))+1)

def r(x):
    return np.exp(-x)

def C(T):
    if T==0:
        return np.NAN
    return e**2/(k*T**2) * np.exp(e/(k*T))*(1+np.exp(e/(k*T)))**(-2)
    
```

```python
X=np.arange(0,N*e+1)
Y=[S_e(i) for i in X]
Z=[S_a(i) for i in X]

plt.rc('figure', figsize=[13,7])
plt.rc('axes', labelsize=15)

plt.plot(X,Y,'o',markersize=7,markeredgewidth=1,markerfacecolor='None' , label='S exact')
plt.plot(X,Z,'m', label='S approximate')
plt.xlabel('E')
plt.ylabel('S')
plt.legend()
plt.grid(True)
plt.show()
```

```python
X1=np.linspace(0,N*e/2,1500)
X2=np.linspace(N*e/2,N*e,1500)
A1=[T(i) for i in X1]
A2=[T(i) for i in X2]

plt.plot(X1,A1,'k')
plt.plot(X2,A2,'k')
plt.xlabel('E')
plt.ylabel('T')
plt.grid(True)
plt.show()
plt.plot(X1,A1,'k')
plt.ylim(-0.1,1)
plt.xlabel('E')
plt.ylabel('T')
plt.grid(True)
plt.show()
```
