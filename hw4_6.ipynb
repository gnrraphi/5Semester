```python
import matplotlib.pyplot as plt
import numpy as np
import numba
```

```python
def pc(r): #plot circle
    t=np.arange(-r-1,r+2)
    n = 0
    for i in t:
        for j in t:
            if i**2+j**2<=r**2:
                n += 1
                plt.plot([i],[j],'kx')
            else:
                plt.plot([i],[j],'mx')
    t=np.linspace(0,2*np.pi,100*r)
    plt.plot([r*np.cos(i) for i in t], [r*np.sin(i) for i in t],'k',alpha=0.6)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.xlabel('x')
    plt.ylabel('y')
    #plt.savefig("hw4_6_1.pdf", bbox_inches='tight')
    plt.show()

@numba.njit()
def ap(r): #aproximate pi
    t=[i-r for i in range(2*r+1)]
    n = 0
    for i in t:
        for j in t:
            if i**2+j**2<=r**2:
                n += 1
    return n

def pa(r_max): #plot aproximate circle area
    t=np.arange(1,r_max+1)
    y1=[ap(r) for r in t]
    y2=[np.pi*r**2 for r in t]
    plt.plot(t,y1,'m', label="Numeric area")
    plt.plot(t,y2,'k', label="Analytic area")
    plt.legend()
    plt.xlabel('R')
    plt.ylabel('area')
    #plt.savefig("hw4_6_2.pdf", bbox_inches='tight')
    plt.show()
    y3=[ap(r)/r**2 for r in t]
    y4=[np.pi for r in t]
    plt.plot(t,y3,'m', label="Numeric \u03A0")
    plt.plot(t,y4,'k', label="Exact \u03A0")
    plt.xlabel('R')
    plt.legend()
    #plt.savefig("hw4_6_3.pdf", bbox_inches='tight')
    plt.show()
    plt.plot(t,[y1[i]-y2[i] for i in range(len(t))])
    plt.show()
                
                
```

```python
plt.rc('figure', figsize=[13,7])

pc(13)
```
