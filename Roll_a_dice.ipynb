```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
def prob_HH(n):
    y=[]
    for i in range(n):
        m = 0
        x = ""
        while(x[-2:]!="HH"):
            r=random.randrange(0,2)
            if r==0:
                x += "H"
            else:
                x += "T"
            m += 1
        y.append(m)
    return np.average(y)

def prob_HT(n):
    y=[]
    for i in range(n):
        m = 0
        x = ""
        while(x[-2:]!="HT"):
            r=random.randrange(0,2)
            if r==0:
                x += "H"
            else:
                x += "T"
            m += 1
        y.append(m)
    return np.average(y)
```

```python
X=np.arange(10,100000,100)
Y_HH=[prob_HH(i) for i in X]
Y_HT=[prob_HT(i) for i in X]

plt.rc('figure', figsize=[13,7])
plt.plot(X,Y_HH,'b')
plt.plot(X,Y_HT,'r')
plt.grid(linestyle='--')
plt.hlines(4, xmin=X[0], xmax=X[-1], alpha=0.8)
plt.hlines(6, xmin=X[0], xmax=X[-1], alpha=0.8)
plt.show()
```
