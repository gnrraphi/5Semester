```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
def prob_HH(n):
    y=[]
    for i in range(n):
        m = 0
        x = ""
        while(x[-2:]!="HH"):
            r=random.randrange(0,2)
            if r==0:
                x += "H"
            else:
                x += "T"
            m += 1
        y.append(m)
    return np.average(y)

def prob_HT(n):
    y=[]
    for i in range(n):
        m = 0
        x = ""
        while(x[-2:]!="HT"):
            r=random.randrange(0,2)
            if r==0:
                x += "H"
            else:
                x += "T"
            m += 1
        y.append(m)
    return np.average(y)
```

```python
X=np.arange(10,100000,100)
Y_HH=[prob_HH(i) for i in X]
Y_HT=[prob_HT(i) for i in X]

plt.rc('figure', figsize=[13,7])
plt.rc('font', size=12)          # controls default text sizes
plt.rc('axes', titlesize=15)     # fontsize of the axes title
plt.rc('axes', labelsize=15)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=12)    # fontsize of the tick labels
plt.rc('ytick', labelsize=12)    # fontsize of the tick labels
plt.rc('legend', fontsize=10)    # legend fontsize
plt.rc('figure', titlesize=12)  # fontsize of the figure title
plt.plot(X,Y_HH,'b',label='Numeric RR')
plt.plot(X,Y_HT,'r',label='Numeric RL')
plt.grid(linestyle='--')
plt.hlines(6, xmin=X[0], xmax=X[-1],colors='k',label='Analytic RR')
plt.hlines(4, xmin=X[0], xmax=X[-1],colors='g',label='Analytic RL')
plt.legend()
plt.xlabel('runs')
plt.ylabel('steps')
#plt.savefig("RR_RL.pdf", bbox_inches='tight')
plt.show()
```
