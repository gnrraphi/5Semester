```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f11(x,y):
    return 1/(y**2+1)

def f21(x,y):
    return 0

def f22(x,y):
    return 1/(y**2+1)
```

```python
def M(x,y):
    return np.linalg.eigh([[f11(x,y),f21(x,y)],[f21(x,y),f22(x,y)]])

def fixed_r(x,y,r):
    D,S = M(x,y)
    t = np.linspace(0, 2*np.pi, 50*int(r))
    X = []
    Y = []
    for i in t:
        x = np.sqrt(r/D[0])*np.cos(i)
        y = np.sqrt(r/D[1])*np.sin(i)
        z = S.dot([x,y])
        X.append(z[0])
        Y.append(z[1])
    return X,Y

def plot_metric(x,y):
    D,S = M(x,y)
    if D[0]<=0 or D[1]<=0:
        print('Metric not positive definite!')
        return 0
    for r in [1,3,5,7,9,11,13,15,17]:
        X,Y = fixed_r(x,y,r)
        plt.plot(X,Y,'b')
    for r in [2,4,6,8,10,12,14,16,18]:
        X,Y = fixed_r(x,y,r)
        plt.plot(X,Y,'g')
    plt.grid(True)
    plt.title('$T_p\mathbb{R}^2$ for p=(%d, %d)'%(x,y))
    plt.xlim(-3, 3)
    plt.ylim(-3, 3)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
```

```python
plt.rc('figure', figsize=[13,7])

plot_metric(0,0)
plot_metric(1,1)
plot_metric(1,-2)
```
